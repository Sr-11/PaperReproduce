```python
from IPython.display import clear_output
import numpy as np
import itertools
import math
import statistics
import matplotlib.pyplot as plt

from distance import rho
from moment_tensor import moment
from sym import sym
from T3estimator import T3estimator
from Jennrich import Jennrich
from generate import generate
from fourier import fourier
```

```python
# Generate theta and Fourier(theta)
d=10
theta=np.empty(d)
cos=math.cos
sin=math.sin
pi=math.pi
exp=math.exp
for k in range(d):
    x=k/d
    theta[k]=k
theta=theta/np.linalg.norm(theta,2)
plt.plot(range(d),theta)
plt.title(r'$\theta$')
plt.show()
theta_hat=fourier(theta)
plt.plot(range(d),np.absolute(theta_hat))
plt.title(r"$|\hat{\theta}|$")
plt.show()
print(min(abs(theta_hat)))
```

```python
# See Jennrich(moment(3,theta))
T3=moment(3,theta)
J=Jennrich(T3,d)
print(J.shape)
print(max(abs(np.imag(J[:,0]))))
plt.plot(range(d),np.real(J[:,0]))
plt.title(r"$Jennrich[T^{(3)}{(\theta)}]$")
```

```python
n=10
sigma=0.01
y=generate(theta,n,sigma)
T=T3estimator(y,sigma)
print(np.linalg.norm(T-T3)/np.linalg.norm(T3))
J=Jennrich(T,d)
plt.plot(range(d),np.real(J[:,0]))
plt.title(r"$\tilde{\theta}$")
```
